## Extraction of Combined Kinematic-Muscular Synergies with Pose Data

This Python script extracts **combined kinematic-muscle synergies** from synchronized kinematic & EMG (Myo and OTB) data for multiple participants, trials and two distinct phases.

- **Global Setup**: Defines participants (P1–P8), trials (1–24), and processing parameters. Configures downsampling from 2000 Hz to 500 Hz.
- **Trial-Condition Mapping**: Maps trials to experimental conditions (grasp type, handle type, weight, etc.).
- **Data Loading & Preprocessing**: Loads kinematic (`*_kin_norm.npy`), EMG (`*_myo.npy`), and OTB (`*_otb.npy`) data. Synchronizes lengths and combines all data into a unified matrix.
- **Downsampling**: Reduces sampling rate using SciPy’s `decimate`.
- **Synergy Extraction**: Decomposes combined data into:
  - **Synergy vectors (`W`)**: Combined kinematic-muscle patterns.
  - **Activations (`C`)**: Temporal dynamics. Enforces non-negativity on EMG components.
- **Evaluation**: Computes **R² scores** to assess reconstruction quality.
- **Visualization**: Plots synergies (OTB, EMG, kinematics) and their temporal activations. Compares original vs. reconstructed signals.
- **Main Workflow**: Processes each participant, trial, and phase: preprocesses, downsamples, extracts synergies, saves results, generates synergy and reconstruction plots, and calculates R² scores.

```python
import os
import numpy as np
from pathlib import Path
import glob
import matplotlib.pyplot as plt
from matplotlib.patches import Patch
from scipy.signal import decimate  # *** Added: Import decimate for downsampling ***

# ---------------------------------------------------------------------
# 0) Global Parameters
# ---------------------------------------------------------------------
BASE_DIR = r"C:\Users\schmi\Documents\Studium\TUM\5. Semester\Masterthesis\Experimental Data"

# List of Participants
PARTICIPANTS = [1, 2]  

PHASES = ["phase1", "phase2"]  
N_SYNERGIES = 5             # Number of synergy components
MAX_ITER = 5000             # Maximum number of MMF iterations
MU_S = 0.1                  # Scaled learning rate
LAMBD_S = 50                # Regularization parameter
NUM_EMG_CHANNELS = 117      # Number of EMG channels in W that must remain non-negative
POSE_COLS = 6               # Number of pose columns 

# *** Added: Downsampling Parameters ***
ORIGINAL_SAMPLING_RATE = 2000.0  # Original sampling rate in Hz
TARGET_SAMPLING_RATE = 1000.0     # Target sampling rate in Hz
DOWNSAMPLE_FACTOR = int(ORIGINAL_SAMPLING_RATE / TARGET_SAMPLING_RATE)
if ORIGINAL_SAMPLING_RATE % TARGET_SAMPLING_RATE != 0:
    raise ValueError("Original sampling rate must be divisible by target sampling rate.")

# ---------------------------------------------------------------------
# 1) Trial => Condition Mapping (Experimental Protocol)
# ---------------------------------------------------------------------
def trial_info(trial_number):
    protocol = {
        1 : ("Precision Grasp (Four Fingers and Thumb)", "Precision Handle", 0.25, "Left Lever",  "No"),
        2 : ("Precision Grasp (Four Fingers and Thumb)", "Precision Handle", 0.25, "Left Lever",  "Yes"),
        3 : ("Lateral Pinch Grasp",                     "Lateral Pinch Handle", 0.25, "Right Lever", "No"),
        4 : ("Lateral Pinch Grasp",                     "Lateral Pinch Handle", 0.25, "Right Lever", "Yes"),
        5 : ("Ball Grasp",                              "Ball Handle", 0.50, "Left Lever",  "No"),
        6 : ("Ball Grasp",                              "Ball Handle", 0.50, "Left Lever",  "Yes"),
        7 : ("Precision Grasp (Thumb and Index)",       "Precision Handle", 0.25, "Front Lever", "No"),
        8 : ("Precision Grasp (Thumb and Index)",       "Precision Handle", 0.25, "Front Lever", "Yes"),
        9 : ("Disc Grip",                               "Disc Handle", 0.50, "Back Lever",  "No"),
        10: ("Disc Grip",                               "Disc Handle", 0.50, "Back Lever",  "Yes"),
        11: ("Power Bar Grasp",                         "Power Bar Handle", 0.50, "Front Lever", "No"),
        12: ("Power Bar Grasp",                         "Power Bar Handle", 0.50, "Front Lever", "Yes"),
        13: ("Precision Grasp (Four Fingers and Thumb)","Precision Handle", 0.25, "Front Lever", "No"),
        14: ("Precision Grasp (Four Fingers and Thumb)","Precision Handle", 0.25, "Front Lever", "Yes"),
        15: ("Lateral Pinch Grasp",                     "Lateral Pinch Handle", 0.25, "Back Lever", "No"),
        16: ("Lateral Pinch Grasp",                     "Lateral Pinch Handle", 0.25, "Back Lever", "Yes"),
        17: ("Ball Grasp",                              "Ball Handle", 0.50, "Front Lever", "No"),
        18: ("Ball Grasp",                              "Ball Handle", 0.50, "Front Lever", "Yes"),
        19: ("Precision Grasp (Thumb and Index)",       "Precision Handle", 0.25, "Back Lever", "No"),
        20: ("Precision Grasp (Thumb and Index)",       "Precision Handle", 0.25, "Back Lever", "Yes"),
        21: ("Disc Grip",                               "Disc Handle", 0.50, "Left Lever",  "No"),
        22: ("Disc Grip",                               "Disc Handle", 0.50, "Left Lever",  "Yes"),
        23: ("Power Bar Grasp",                         "Power Bar Handle", 0.50, "Right Lever","No"),
        24: ("Power Bar Grasp",                         "Power Bar Handle", 0.50, "Right Lever","Yes"),
    }
    if trial_number not in protocol:
        return None
    tup = protocol[trial_number]
    return {
        'grasp_type':  tup[0],
        'handle_type': tup[1],
        'weight_kg':   tup[2],
        'lever_side':  tup[3],
        'knowledge':   tup[4]
    }

# ---------------------------------------------------------------------
# 2) Data Loading & Concatenation (No Cropping, include pose data)
# ---------------------------------------------------------------------
def load_phase_data(participant_dir, trial_idx, phase_name):
    """
    Loads kinematic (kin), Myo, and OTB data for a given participant, trial, and phase.
    1) Crops each array to the same minimum length (n_min).
    2) Includes all columns from the kinematic array (both hand and pose).
       Normalizes only the hand columns to [-1, 1].
    3) Concatenates otb, myo, normalized hand, and pose horizontally => 'combined'.

    Returns:
      {
         'combined':    np.ndarray (n_samples, n_features),
         'kin_hand_indices': (start_col, end_col),
         'kin_pose_indices': (start_col, end_col),
         'myo_indices': (start_col, end_col),
         'otb_indices': (start_col, end_col)
      }
    or None if something is missing or if there are <5 samples.
    """
    sync_phase_dir = os.path.join(participant_dir, "Synchronized Data split in Phases")

    kin_file = os.path.join(sync_phase_dir, f"match_{trial_idx:02d}_{phase_name}_kin_norm.npy")
    myo_file = os.path.join(sync_phase_dir, f"match_{trial_idx:02d}_{phase_name}_myo.npy")
    otb_file = os.path.join(sync_phase_dir, f"match_{trial_idx:02d}_{phase_name}_otb.npy")

    # Check existence
    if not (os.path.exists(kin_file) and os.path.exists(myo_file) and os.path.exists(otb_file)):
        print(f"[DEBUG] Missing file(s) for trial={trial_idx}, phase={phase_name}")
        return None

    # Load
    kin = np.load(kin_file)
    myo = np.load(myo_file)
    otb = np.load(otb_file)

    # Crop each to same min length
    n_min = min(kin.shape[0], myo.shape[0], otb.shape[0])
    if n_min < 5:
        print(f"[DEBUG] Not enough frames for trial={trial_idx}, phase={phase_name} => {n_min} frames.")
        return None

    kin_c = kin[:n_min]
    myo_c = myo[:n_min]
    otb_c = otb[:n_min]

    # -----------------------------------------------------------------
    # **Removed the exclusion of pose data and include all kin data**
    # -----------------------------------------------------------------
    if POSE_COLS > 0:
        hand_only = kin_c[:, : (kin_c.shape[1] - POSE_COLS)]
        pose_only = kin_c[:, (kin_c.shape[1] - POSE_COLS):]
    else:
        # If POSE_COLS=0, all kin data is considered as hand
        hand_only = kin_c
        pose_only = np.empty((kin_c.shape[0], 0))  # Empty array for pose

    # Normalize hand columns to [-1, 1]
    hand_norm = hand_only
    pose_norm = pose_only

    # Concatenate otb, myo, normalized hand, and pose
    combined = np.hstack([otb_c, myo_c, hand_norm, pose_norm])

    # Update indices to reflect new concatenation order
    otb_cols = otb_c.shape[1]
    myo_cols = myo_c.shape[1]
    kin_hand_cols = hand_norm.shape[1]
    kin_pose_cols = pose_norm.shape[1]

    otb_start, otb_end = (0, otb_cols)
    myo_start, myo_end = (otb_end, otb_end + myo_cols)
    kin_hand_start, kin_hand_end = (myo_end, myo_end + kin_hand_cols)
    kin_pose_start, kin_pose_end = (kin_hand_end, kin_hand_end + kin_pose_cols)

    # Debug prints
    print("[DEBUG] Loading data for trial={}, phase={}".format(trial_idx, phase_name))
    print(f"    KIN Hand shape (normalized) = {hand_norm.shape}, columns => [{kin_hand_start}..{kin_hand_end-1}] (normalized -1..1)")
    print(f"    KIN Pose shape = {pose_norm.shape}, columns => [{kin_pose_start}..{kin_pose_end-1}]")
    print(f"    MYO shape (cropped) = {myo_c.shape}, columns => [{myo_start}..{myo_end-1}]")
    print(f"    OTB shape (cropped) = {otb_c.shape}, columns => [{otb_start}..{otb_end-1}]")
    print(f"    => Combined shape = {combined.shape}")

    return {
        'combined': combined,
        'kin_hand_indices': (kin_hand_start, kin_hand_end),
        'kin_pose_indices': (kin_pose_start, kin_pose_end),
        'myo_indices': (myo_start, myo_end),
        'otb_indices': (otb_start, otb_end)
    }

def downsample_data(data_mat, factor):
    """
    Downsamples the data matrix by the specified factor along the time axis (axis=0).
    Args:
        data_mat (np.ndarray): The data matrix to downsample (n_samples, n_features).
        factor (int): The downsampling factor.
    Returns:
        np.ndarray: The downsampled data matrix.
    """
    # Using zero_phase=True to prevent phase shift
    return decimate(data_mat, factor, axis=0, zero_phase=True)

# ---------------------------------------------------------------------
# 3) Mixed Matrix Factorization Algorithm
# ---------------------------------------------------------------------
def initialize_matrices(X, n_synergies):
    """Initialize W and C with small random values, shape:
       W => (vars x n_synergies),  C => (n_synergies x samples)."""
    np.random.seed(0)  # reproducibility
    n_vars, n_samples = X.shape

    # Initialize W and C
    W = np.random.rand(n_vars, n_synergies) * 0.01
    C = np.random.rand(n_synergies, n_samples) * 0.01
    return W, C

def update_mmf(X, W, C, mu_W, mu_C, lambd, k):
    """
    One iteration of gradient-based update:
      X => (vars x samples)
      W => (vars x synergies)
      C => (synergies x samples)
      k => # of EMG rows in W that must be non-negative
    """
    # Compute updates
    W_update = mu_W * ((X @ C.T) - (W @ (C @ C.T)) - lambd * W)
    C_update = mu_C * ((W.T @ X) - (W.T @ W @ C))

    # Gradient clipping
    W_update = np.clip(W_update, -0.01, 0.01)
    C_update = np.clip(C_update, -0.01, 0.01)

    # Apply updates
    W += W_update
    C += C_update

    # Enforce non-negativity for the EMG portion of W => rows [0..k-1]
    W[:k, :] = np.maximum(W[:k, :], 0)

    return W, C

def mixed_matrix_factorization(
    data_matrix, 
    n_synergies=5, 
    k=117,              # e.g., # of EMG channels
    mu_s=0.1, 
    lambd_s=50, 
    max_iter=5000
):
    """
    Full MMF pipeline:
      1) Transpose data_matrix => (vars x samples).
      2) Initialize W, C.
      3) Iteratively update W, C, track cost.
      4) Return final W, C (W => (vars x n_synergies), C => (n_synergies x samples)).

    - Early stopping: stop as soon as the cost function is below 1.
    """
    # Prepare X => (n_features, n_samples)
    X = data_matrix.T
    n_vars, n_samples = X.shape

    # Initialize W, C
    np.random.seed(0) 
    W = np.random.rand(n_vars, n_synergies) * 0.01
    C = np.random.rand(n_synergies, n_samples) * 0.01

    # Scale the learning rate & lambda
    normX = np.linalg.norm(X)
    mu_W = mu_s / normX
    mu_C = mu_s / normX
    lambd = lambd_s / (n_vars * n_synergies)

    cost_history = []

    for iteration in range(max_iter):
        # ---------------------------
        # Gradient-based update
        # ---------------------------
        # 1) Compute W_update
        W_update = mu_W * ((X @ C.T) - (W @ (C @ C.T)) - lambd * W)
        # 2) Compute C_update
        C_update = mu_C * ((W.T @ X) - (W.T @ W @ C))

        # 3) Gradient clipping
        W_update = np.clip(W_update, -0.01, 0.01)
        C_update = np.clip(C_update, -0.01, 0.01)

        # 4) Apply updates
        W += W_update
        C += C_update

        # 5) Enforce non-negativity for EMG portion of W => rows [0..k-1]
        W[:k, :] = np.maximum(W[:k, :], 0)

        # ---------------------------
        # Cost function
        # ---------------------------
        residual = X - (W @ C)
        cost = 0.5 * np.linalg.norm(residual, 'fro')**2 + 0.5 * lambd * np.linalg.norm(W, 'fro')**2
        cost_history.append(cost)

        # ---------------------------
        # Early stopping criterion
        # ---------------------------
        if cost < 1:
            print(f"Stopping early at iteration {iteration} with cost={cost:.6f} < 1")
            break

        # ---------------------------
        # Debug printing & quick plot
        # ---------------------------
        if iteration % 500 == 0:
            print(f"Iteration {iteration}, Cost: {cost:.6f}")
            print(f"W matrix (first few values) for synergy 0 => {W[:5, 0]}")
            print(f"C matrix (first few values) for synergy 0 => {C[0, :5]}")

            plt.figure(figsize=(10, 4))

            # *** Modified: Use TARGET_SAMPLING_RATE ***
            sampling_rate = TARGET_SAMPLING_RATE  # *** Changed from 30.0 to TARGET_SAMPLING_RATE ***

            # Bar plot => W[:, 0]
            plt.subplot(1, 2, 1)
            plt.bar(np.arange(W.shape[0]), W[:, 0], color='#4F81BD', label='Synergy 1')
            plt.title('W Matrix First Synergy (Bar Plot)')
            plt.xlabel('Variables')
            plt.ylabel('Synergy Component Value')
            plt.legend()

            time_axis = np.arange(C.shape[1]) / sampling_rate

            plt.subplot(1, 2, 2)
            plt.plot(time_axis, C[0, :], label='Coefficient 1', color='#F4A32F')
            plt.title('C Matrix First Coefficient (Line Plot)')
            plt.xlabel('Time (s)')
            plt.ylabel('Activation')
            plt.legend()

            plt.tight_layout()
            plt.show()

    # Final cost plot if you like
    plt.figure()
    plt.plot(cost_history, '-o', markersize=3)
    plt.title("MMF Cost Function Convergence")
    plt.xlabel("Iteration")
    plt.ylabel("Cost")
    plt.grid(True)
    plt.show()

    return W, C

# ---------------------------------------------------------------------
# 4) Wrapper for synergy extraction
# ---------------------------------------------------------------------
def mmf_extract_synergies(data_matrix, n_synergies=5):
    """
    data_matrix: shape (n_samples, n_features).
    Returns synergy vectors W => (n_features, n_synergies)
            synergy activations H => (n_synergies, n_samples)
    """
    W_raw, C_raw = mixed_matrix_factorization(
        data_matrix, 
        n_synergies=n_synergies,
        k=NUM_EMG_CHANNELS,
        mu_s=MU_S,
        lambd_s=LAMBD_S,
        max_iter=MAX_ITER
    )
    return W_raw, C_raw

# ---------------------------------------------------------------------
# 5) R2 Calculation
# ---------------------------------------------------------------------
def r2_score_matrix(original_2d, reconstructed_2d):
    """
    Computes a single R^2 value across all elements of original_2d and reconstructed_2d.
    Both must have the same shape => (n_features, n_samples).
    """
    # Flatten so we compare all data points at once
    orig_flat = original_2d.flatten()
    reconst_flat = reconstructed_2d.flatten()

    ss_res = np.sum((orig_flat - reconst_flat) ** 2)
    ss_tot = np.sum((orig_flat - np.mean(orig_flat)) ** 2)
    return 1.0 - (ss_res / ss_tot)

# ---------------------------------------------------------------------
# 6) Plotting Function
# ---------------------------------------------------------------------
def plot_and_save_synergies(W, H, synergy_out_dir, prefix, index_dict):
    """
    Plots synergies with the specified coloring:
      - Kinematic Hand => orange
      - Kinematic Pose => red
      - Myo => pink
      - OTB => blue
    """

    # Validate inputs
    n_features, n_synergies = W.shape
    if H.shape[0] != n_synergies:
        raise ValueError("Number of synergies in W and H must match.")
    
    # Prepare color array
    colors = [""] * n_features
    kin_hand_start, kin_hand_end = index_dict['kin_hand_indices']
    kin_pose_start, kin_pose_end = index_dict['kin_pose_indices']
    myo_start, myo_end = index_dict['myo_indices']
    otb_start, otb_end = index_dict['otb_indices']

    for i in range(kin_hand_start, kin_hand_end):
        colors[i] = 'orange'
    for i in range(kin_pose_start, kin_pose_end):
        colors[i] = 'red'
    for i in range(myo_start, myo_end):
        colors[i] = 'pink'
    for i in range(otb_start, otb_end):
        colors[i] = '#4F81BD'  # Corrected color code for OTB

    # Create output directory if needed
    os.makedirs(synergy_out_dir, exist_ok=True)

    # Plot
    fig_height = 4 * n_synergies
    fig, axes = plt.subplots(n_synergies, 2, figsize=(20, fig_height))
    if n_synergies == 1:
        axes = np.array([axes])  # Ensure 2D indexing

    for i in range(n_synergies):
        # Bar Plot (W[:, i])
        ax_bar = axes[i, 0]
        ax_bar.bar(np.arange(n_features), W[:, i], color=colors, alpha=0.7, width=1.0)
        ax_bar.set_title(f"Synergy {i+1} (W column {i})", fontsize=14)
        ax_bar.set_xlabel("Features", fontsize=12)
        ax_bar.set_ylabel("Amplitude", fontsize=12)

        # Add text labels for each section
        if kin_pose_end > kin_hand_end:
            ax_bar.text((otb_start + otb_end)/2, ax_bar.get_ylim()[1]*0.95,
                        'OTB', ha='center', va='top', fontsize=12, color='#4F81BD')
            ax_bar.text((myo_start + myo_end)/2, ax_bar.get_ylim()[1]*0.95,
                        'Myo', ha='center', va='top', fontsize=12, color='pink')
            ax_bar.text((kin_hand_start + kin_hand_end)/2, ax_bar.get_ylim()[1]*0.95,
                        'Kinematic Hand', ha='center', va='top', fontsize=12, color='orange')
            ax_bar.text((kin_pose_start + kin_pose_end)/2, ax_bar.get_ylim()[1]*0.95,
                        'Kinematic Pose', ha='center', va='top', fontsize=12, color='red')
        else:
            ax_bar.text((otb_start + otb_end)/2, ax_bar.get_ylim()[1]*0.95,
                        'OTB', ha='center', va='top', fontsize=12, color='#4F81BD')
            ax_bar.text((myo_start + myo_end)/2, ax_bar.get_ylim()[1]*0.95,
                        'Myo', ha='center', va='top', fontsize=12, color='pink')
            ax_bar.text((kin_hand_start + kin_hand_end)/2, ax_bar.get_ylim()[1]*0.95,
                        'Kinematic', ha='center', va='top', fontsize=12, color='orange')

        # Remove x-ticks for clarity
        ax_bar.set_xticks([])

        # Create custom legend
        legend_elements = [
            Patch(facecolor='orange', label='Kinematic Hand'),
            Patch(facecolor='red', label='Kinematic Pose'),
            Patch(facecolor='pink', label='Myo'),
            Patch(facecolor='#4F81BD', label='OTB')  # Corrected color code for OTB
        ]
        ax_bar.legend(handles=legend_elements, loc='upper right')

        # Line Plot (H[i, :])
        ax_line = axes[i, 1]
        # *** Modified: Use TARGET_SAMPLING_RATE ***
        sampling_rate = TARGET_SAMPLING_RATE  # *** Changed from 30.0 to TARGET_SAMPLING_RATE ***
        time_axis = np.arange(H.shape[1]) / sampling_rate

        ax_line.plot(time_axis, H[i, :], label='Activation {}'.format(i+1), color='green', linewidth=1.5)
        ax_line.set_title(f"Activation {i+1} (H row {i})", fontsize=14)
        ax_line.set_xlabel('Time (s)', fontsize=12)
        ax_line.set_ylabel('Activation', fontsize=12)
        ax_line.grid(True, which='both', linestyle='--', linewidth=0.5)

    plt.tight_layout()
    out_filename = f"{prefix}_synergies.png"
    out_path = os.path.join(synergy_out_dir, out_filename)
    plt.savefig(out_path, dpi=150)
    plt.close(fig)

    print(f"[INFO] Synergy plots saved to {out_path}")

# ---------------------------------------------------------------------
# 7) Main Script
# ---------------------------------------------------------------------
def main():
    for pid in PARTICIPANTS:
        participant_str = f"P({pid})"
        participant_dir = os.path.join(BASE_DIR, participant_str)
        if not os.path.isdir(participant_dir):
            print(f"[WARN] Participant folder not found: {participant_str}")
            continue

        synergy_out_dir = os.path.join(participant_dir, "Extracted Synergies")
        os.makedirs(synergy_out_dir, exist_ok=True)

        for trial_idx in range(1, 25):
            info_dict = trial_info(trial_idx)
            if info_dict is None:
                continue

            for phase_name in PHASES:
                data_dict = load_phase_data(participant_dir, trial_idx, phase_name)
                if data_dict is None:
                    print(f"[INFO] Missing or invalid data for {participant_str} trial {trial_idx}, {phase_name}")
                    continue

                data_mat = data_dict['combined']  # shape (n_samples, n_features)

                # *** Added: Downsample the data ***
                data_mat_ds = downsample_data(data_mat, DOWNSAMPLE_FACTOR)
                print(f"[DEBUG] Downsampled data from {data_mat.shape[0]} to {data_mat_ds.shape[0]} samples.")

                W, H = mmf_extract_synergies(data_mat_ds, n_synergies=N_SYNERGIES)

                # Save synergy factors
                prefix = f"trial_{trial_idx:02d}_{phase_name}"
                W_file = os.path.join(synergy_out_dir, prefix + "_W.npy")
                H_file = os.path.join(synergy_out_dir, prefix + "_H.npy")
                np.save(W_file, W)
                np.save(H_file, H)

                cond_str = info_dict['knowledge']
                print(f"[OK] Extracted synergies for {participant_str}, trial {trial_idx} ({cond_str}), {phase_name}")
                print("[DEBUG] W matrix shape:", W.shape)
                print("[DEBUG] H matrix shape:", H.shape)

                # ----------------------------------------------------------
                # Plot the synergy vectors & activations
                # ----------------------------------------------------------
                plot_and_save_synergies(
                    W, H, synergy_out_dir, prefix,
                    {
                        'kin_hand_indices': data_dict['kin_hand_indices'],
                        'kin_pose_indices': data_dict['kin_pose_indices'],
                        'myo_indices': data_dict['myo_indices'],
                        'otb_indices': data_dict['otb_indices'],
                    }
                )

                # ----------------------------------------------------------
                # After synergy extraction: Reconstruction & R^2
                # ----------------------------------------------------------
                # Original data in shape (n_samples, n_features)
                # But our synergy factorization used X = data_matrix^T => (n_features, n_samples).
                # Reconstruct => (n_features, n_samples), compare to original X
                X_original = data_mat_ds.T              # (n_features, n_samples)
                X_reconstructed = W @ H              # (n_features, n_samples)

                # Compute R^2 for the entire matrix
                r2_val = r2_score_matrix(X_original, X_reconstructed)
                print(f"Combined Data Reconstruction R^2: {r2_val:.4f}")

                # Choose channel indices to visualize
                channel_indices_to_plot = [5, 130]

                # Create a figure for reconstruction plots
                fig, axs = plt.subplots(len(channel_indices_to_plot), 1, 
                                        figsize=(12, 4 * len(channel_indices_to_plot)))

                # If there's only one channel index to plot, make `axs` a list for convenience
                if len(channel_indices_to_plot) == 1:
                    axs = [axs]

                for ax_idx, ch_idx in enumerate(channel_indices_to_plot):
                    ax = axs[ax_idx]
                    if ch_idx < X_original.shape[0]:
                        ax.plot(X_original[ch_idx, :],
                                label=f'Original (ch={ch_idx})', alpha=0.7)
                        ax.plot(X_reconstructed[ch_idx, :],
                                label=f'Reconstructed (ch={ch_idx})', linestyle='--')
                        ax.set_title(f"Original vs. Reconstructed (Channel {ch_idx})")
                        ax.set_xlabel('Time (Samples)')
                        ax.set_ylabel('Amplitude')
                        ax.legend()
                    else:
                        ax.set_title(f"[INFO] Channel {ch_idx} out of range.")
                        ax.set_xticks([])
                        ax.set_yticks([])

                # *** Modified: Use TARGET_SAMPLING_RATE for the suptitle ***
                fig.suptitle(f"Reconstruction Plot (R^2 = {r2_val:.4f}) - Sampling Rate: {TARGET_SAMPLING_RATE} Hz", fontsize=16)

                plt.tight_layout(rect=[0, 0, 1, 0.96])  # Leave space for suptitle
                recon_plot_file = os.path.join(synergy_out_dir, f"{prefix}_reconstruction_plot.png")
                plt.savefig(recon_plot_file, dpi=150)
                plt.close(fig)


        print(f"[INFO] Synergy extraction complete for {participant_str}.\n"
              f"Plots and synergy data in => {synergy_out_dir}")

if __name__ == "__main__":
    main()
```
