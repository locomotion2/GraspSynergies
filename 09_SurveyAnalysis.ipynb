## Participant Survey Data Analysis and Reliability Assessment

This script loads and processes participant survey responses related to EMG studies. It performs the following:

1. **Data Loading:** Traverses directories to load survey data from `.npy` files.
2. **Data Processing:** Organizes responses, handles reverse-coded items, and computes construct scores.
3. **Reliability Analysis:** Calculates Cronbach's Alpha, Ordinal Alpha, McDonald's Omega and Consistency Scores to assess internal consistency and response reliability.
4. **Factor Analysis:** Conducts PCA to identify underlying factors and visualizes loadings.
5. **Statistical Testing:** Performs Wilcoxon tests to compare constructs across conditions.
6. **Visualization:** Generates heatmaps, boxplots, swarmplots, and consistency tables for data insights.
7. **Integration with R:** Utilizes the `polycor` R package via `rpy2` for polychoric correlations.

**Dependencies:**
- Python libraries: numpy, pandas, scipy, matplotlib, seaborn, scikit-learn, factor_analyzer, pingouin, rpy2
- R package: polycor

```python
import os
import numpy as np
import pandas as pd
import warnings
from scipy.stats import wilcoxon, spearmanr
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from factor_analyzer import FactorAnalyzer
import pingouin as pg  
import rpy2.robjects as ro
from rpy2.robjects import pandas2ri
from rpy2.robjects.packages import importr
from matplotlib.table import Table
from scipy.stats import wilcoxon

# Import polycor globally
polycor = importr('polycor')
pandas2ri.activate()

# Base directory where participant data is stored
base_path = r"C:\Users\schmi\Documents\Studium\TUM\5. Semester\Masterthesis\Experimental Data"

def load_participant_data(base_path):
    """Load survey responses from all participant files."""
    data = []
    participant_dirs = [d for d in os.listdir(base_path) if os.path.isdir(os.path.join(base_path, d))]

    for participant_dir in participant_dirs:
        participant_path = os.path.join(base_path, participant_dir)
        files = [f for f in os.listdir(participant_path) if f.endswith('_responses.npy')]
        
        for file in files:
            file_path = os.path.join(participant_path, file)
            try:
                participant_data = np.load(file_path, allow_pickle=True).item()
                details = participant_data['details']
                responses = participant_data['responses']
                data.append({
                    'participant_id': details['participant_id'],
                    'responses': responses
                })
            except Exception as e:
                print(f"Error loading file {file_path}: {e}")
    return data

def process_data(data):
    """Process the raw data into DataFrames for original and processed (reverse-coded) responses."""
    Statements = {
        1: "I consciously think about the location of the weight before grasping.",
        2: "I feel more muscle strain when I do not know the weight location.",
        3: "I feel more comfortable with my grip when I know the weight location.",
        4: "I do not adjust my grasp on the handle based on the weight location.",  # Reverse-coded
        5: "I find it easy to lift the object without tilting when I do not know the weight location.",
        6: "I feel less muscle strain when I know the weight location.",  # Reverse-coded
        7: "I find it difficult to lift the object without tilting when I know the weight location.",  # Reverse-coded
        8: "I feel uncomfortable with my grip on the object when I know the weight location.",  # Reverse-coded
        9: "I do not consciously think about the location of the weight before grasping.",  # Reverse-coded
        10: "My muscles feel more fatigued when I know the weight location.",  # Reverse-coded
        11: "My grip feels just as secure when I do not know the weight location.",
        12: "I find it easier to lift the object parallel to the table when I know the weight location.",
        13: "I adjust my grasp on the handle based on the weight location.",
        14: "My muscles feel less fatigued when I know the weight location.",
        15: "I find it difficult to lift the object without tilting when I do not know the weight location.",  # Reverse-coded
        16: "My grip feels safer when I know the weight location.",
        17: "I regularly engage in strength training exercises for my arm muscles."
    }

    data_rows = []
    for participant in data:
        participant_id = participant['participant_id']
        responses = participant['responses']
        if len(responses) != 17:
            print(f"Participant {participant_id} has {len(responses)} responses instead of 17.")
            continue
        row = {'participant_id': participant_id}
        for i, resp in enumerate(responses):
            row[f'Q{i+1}'] = resp
        data_rows.append(row)
    
    df_original = pd.DataFrame(data_rows)

    df_constructs = df_original.copy()
    df_consistency = df_original.copy()

    df_Statements = pd.DataFrame.from_dict(Statements, orient='index', columns=['Statement_Text'])
    df_Statements.index = [f'Q{i}' for i in df_Statements.index]

    reversed_items_constructs = [4, 6, 7, 8, 9, 10, 15]  
    for item in reversed_items_constructs:
        df_constructs[f'Q{item}'] = 8 - df_constructs[f'Q{item}']
    
    reversed_items_consistency = [4, 6, 7, 8, 9, 10, 11, 15]  
    for item in reversed_items_consistency:
        df_consistency[f'Q{item}'] = 8 - df_consistency[f'Q{item}']

    return df_constructs, df_consistency, df_original, df_Statements

def compute_constructs(df_constructs):
    """Compute constructs from survey Statements."""
    constructs = {
        'Awareness': [1, 9],
        'Muscle Strain': [2, 6], 
        'Fatigue': [10, 14],
        'Grip Comfort': [3, 8, 11, 16],
        'Adjust Grasp': [4, 13],
        'Task Ease': [5, 7, 12, 15],
        'Strength Training': [17]  
    }
    df_constructs_mean = df_constructs.copy()
    for construct, Statements in constructs.items():
        df_constructs_mean[construct] = df_constructs[[f'Q{i}' for i in Statements]].mean(axis=1)
    return df_constructs_mean, constructs

def compute_consistency_score(df_consistency, reverse_coded_pairs, exclude_pairs, output_dir=None):
    if exclude_pairs is None:
        exclude_pairs = []
    filtered_pairs = [pair for pair in reverse_coded_pairs if pair not in exclude_pairs]
    participant_scores = []
    print("Debugging Consistency Scores for Each Participant:")
    print("----------------------------------------------------")
    
    for index, row in df_consistency.iterrows():
        participant_id = row['participant_id']
        total_distance = 0
        max_distance = 6 * len(filtered_pairs)
        
        print(f"\nParticipant {participant_id}:")
        
        for pos_item, rev_item in filtered_pairs:
            positive_value = row[pos_item]
            reverse_value = row[rev_item]
            print(f"Participant {participant_id} | {pos_item}: {positive_value}, {rev_item}: {reverse_value}")
            distance = abs(positive_value - reverse_value)
            print(f"  Pair: {pos_item} vs {rev_item}")
            print(f"    {pos_item} value: {positive_value}")
            print(f"    {rev_item} value: {reverse_value}")
            print(f"    Distance: {distance}")
            total_distance += distance
        
        consistency_score = total_distance / max_distance
        participant_scores.append(consistency_score)
        print(f"  Total Distance: {total_distance}")
        print(f"  Normalized Consistency Score: {consistency_score}")
    
    df_consistency_scores = pd.DataFrame({
        'Participant_ID': df_consistency['participant_id'],
        'Consistency_Score': participant_scores
    })
    
    avg_consistency_scores = df_consistency_scores['Consistency_Score'].mean()
    variance_consistency = df_consistency_scores['Consistency_Score'].var()
    
    print("\nSummary of Consistency Scores:")
    print(f"  Average Consistency Score: {avg_consistency_scores:.4f}")
    print(f"  Variance in Consistency Scores: {variance_consistency:.4f}")
    
    return df_consistency_scores, avg_consistency_scores, variance_consistency

def plot_consistency_table(df_consistency_scores, avg_consistency, var_consistency, output_dir):
    df_consistency_scores["Consistency_Score"] = df_consistency_scores["Consistency_Score"].round(3)
    data = df_consistency_scores.values.tolist()
    columns = ["Participant_ID", "Consistency_Score"]
    data.append(["Average", f"{avg_consistency:.3f}"])
    data.append(["Variance", f"{var_consistency:.3f}"])
    fig, ax = plt.subplots(figsize=(6, len(data) * 0.5))
    ax.axis('tight')
    ax.axis('off')
    table = ax.table(
        cellText=data,
        colLabels=columns,
        loc='center',
        cellLoc='center',
        edges='horizontal'
    )
    table.auto_set_font_size(False)
    table.set_fontsize(10)
    table.auto_set_column_width(col=list(range(len(columns))))
    for key, cell in table.get_celld().items():
        cell.set_fontsize(10)
        cell.set_facecolor("#f9f9f9")
        if key[0] == 0:
            cell.set_fontsize(11)
            cell.set_facecolor("#d9d9d9")
    plt.title("Consistency Scores by Participant", fontsize=14, fontname="Arial", pad=20)
    table_filename = os.path.join(output_dir, "Consistency_Scores_Table.png")
    plt.savefig(table_filename, bbox_inches='tight')
    print(f"Consistency table saved to {table_filename}")
    plt.show()

def compute_polychoric_matrix(itemscores):
    r_itemscores = pandas2ri.py2rpy(itemscores)
    result = polycor.hetcor(r_itemscores)
    polychoric_matrix = result.rx2('correlations')
    return np.array(polychoric_matrix)

def ordinal_alpha(polychoric_matrix, itemscores):
    k = itemscores.shape[1]
    total_var = np.sum(polychoric_matrix)
    alpha = (k / (k - 1)) * (1 - (np.sum(np.diag(polychoric_matrix)) / total_var))
    return alpha

def reliability_analysis_pg(df_constructs, constructs):
    alphas = {}
    for construct, Statements in constructs.items():
        if len(Statements) > 1:
            itemscores = df_constructs[[f'Q{i}' for i in Statements]]
            alpha, _ = pg.cronbach_alpha(itemscores)
            alphas[construct] = alpha
            print(f"Cronbach's alpha for {construct}: {alpha:.2f}")
    return alphas

def mcdonald_omega(itemscores):
    fa = FactorAnalyzer(n_factors=1, method='minres', rotation=None)
    fa.fit(itemscores)
    loadings = fa.loadings_.flatten()
    uniqueness = fa.get_uniquenesses()
    numerator = (np.sum(loadings))**2
    denominator = numerator + np.sum(uniqueness)
    omega = numerator / denominator
    return omega

def factor_analysis(df_constructs, output_dir):
    Statement_cols = [f'Q{i}' for i in range(1, 18)]
    X = df_constructs[Statement_cols]
    scaler = StandardScaler()
    X_scaled = scaler.fit_transform(X)
    pca = PCA(n_components=None)
    pca.fit(X_scaled)
    plt.figure(figsize=(14, 6))
    plt.subplot(1, 2, 1)
    plt.bar(range(1, len(pca.explained_variance_ratio_) + 1),
            pca.explained_variance_ratio_, alpha=0.7, align='center', color='skyblue')
    plt.title('Scree Plot', fontsize=16)
    plt.xlabel('Principal Components', fontsize=14)
    plt.ylabel('Explained Variance Ratio', fontsize=14)
    plt.xticks(range(1, len(pca.explained_variance_ratio_) + 1), fontsize=12)
    plt.yticks(fontsize=12)
    plt.subplot(1, 2, 2)
    plt.plot(range(1, len(pca.explained_variance_ratio_) + 1),
             np.cumsum(pca.explained_variance_ratio_), marker='o', linestyle='--', color='green')
    plt.title('Cumulative Explained Variance', fontsize=16)
    plt.xlabel('Number of Components', fontsize=14)
    plt.ylabel('Cumulative Explained Variance Ratio', fontsize=14)
    plt.xticks(range(1, len(pca.explained_variance_ratio_) + 1), fontsize=12)
    plt.yticks(fontsize=12)
    plt.grid(True)
    plt.tight_layout()
    scree_plot_path = os.path.join(output_dir, "PCA_Scree_Plot.png")
    cumulative_plot_path = os.path.join(output_dir, "PCA_Cumulative_Explained_Variance.png")
    plt.savefig(scree_plot_path, bbox_inches='tight')
    plt.savefig(cumulative_plot_path, bbox_inches='tight')
    print(f"Scree plot saved to {scree_plot_path}")
    print(f"Cumulative explained variance plot saved to {cumulative_plot_path}")
    plt.show()
    loadings = pd.DataFrame(pca.components_.T,
                            columns=[f'PC{i+1}' for i in range(len(pca.components_))],
                            index=Statement_cols)
    print("PCA Component Loadings:")
    print(loadings)
    plt.figure(figsize=(12, 8))
    sns.heatmap(loadings, annot=True, cmap='Spectral', fmt='.2f', linewidths=.5)
    plt.title('PCA Component Loadings Heatmap', fontsize=16)
    plt.ylabel('Survey Statements', fontsize=14)
    plt.xlabel('Principal Components', fontsize=14)
    plt.xticks(fontsize=12)
    plt.yticks(fontsize=12)
    plt.tight_layout()
    loadings_heatmap_path = os.path.join(output_dir, "PCA_Loadings_Heatmap.png")
    plt.savefig(loadings_heatmap_path, bbox_inches='tight')
    print(f"PCA loadings heatmap saved to {loadings_heatmap_path}")
    plt.show()
    return pca, loadings

def analyze_data(df_constructs, constructs, output_dir):
    print("\nReliability Analysis:")
    alphas = reliability_analysis_pg(df_constructs, constructs)
    print("\nFactor Analysis:")
    pca, loadings = factor_analysis(df_constructs, output_dir)
    print("\nCorrelation Analysis:")
    correlation_analysis(df_constructs, constructs, output_dir)

    condition_items = {
        'Muscle Strain': {
            'Known': [6],
            'Unknown': [2]
        },
        'Grip Comfort': {
            'Known': [16],
            'Unknown': [11]
        },
        'Task Ease': {
            'Known': [7, 12],
            'Unknown': [5, 15]
        }
    }

    for construct in ['Muscle Strain', 'Grip Comfort', 'Task Ease']:
        known_items = condition_items[construct]['Known']
        unknown_items = condition_items[construct]['Unknown']
        df_constructs[f'{construct}_Known'] = df_constructs[[f'Q{i}' for i in known_items]].mean(axis=1)
        df_constructs[f'{construct}_Unknown'] = df_constructs[[f'Q{i}' for i in unknown_items]].mean(axis=1)

        with warnings.catch_warnings():
            warnings.simplefilter("ignore")
            stat, p_value = wilcoxon(df_constructs[f'{construct}_Known'], df_constructs[f'{construct}_Unknown'])
        print(f"\nWilcoxon Test between {construct} in Known and Unknown conditions:")
        print(f"Statistic: {stat:.4f}, p-value: {p_value:.4f}")
        
        if p_value < 0.001:
            significance_text = "p < 0.001 ***"
        elif p_value < 0.01:
            significance_text = "p < 0.01 **"
        elif p_value < 0.05:
            significance_text = "p < 0.05 *"
        else:
            significance_text = f"p = {p_value:.3f} ns"

        if p_value < 0.05:
            print(f"Result is statistically significant ({significance_text}).")
        else:
            print(f"Result is not statistically significant ({significance_text}).")

        fig, axes = plt.subplots(1, 2, figsize=(16, 6))
        ax_box = axes[0]
        data_melted = df_constructs[[f'{construct}_Known', f'{construct}_Unknown']].melt(var_name='Condition', value_name='Score')
        data_melted['Condition'] = data_melted['Condition'].str.replace(f'{construct}_', '')
        sns.boxplot(x='Condition', y='Score', data=data_melted, palette="Set2", ax=ax_box)
        sns.swarmplot(x='Condition', y='Score', data=data_melted, color=".25", size=7, ax=ax_box)
        ax_box.set_title(f'{construct.replace("_", " ")}: Known vs Unknown', fontsize=16)
        ax_box.set_ylabel('Rating (1-7)', fontsize=14)
        ax_box.set_xlabel('')
        ax_box.tick_params(axis='x', labelsize=12)

        ax_diff = axes[1]
        for i in range(len(df_constructs)):
            ax_diff.plot(['Known', 'Unknown'], 
                        [df_constructs[f'{construct}_Known'].iloc[i], df_constructs[f'{construct}_Unknown'].iloc[i]],
                        color='gray', linestyle='-', marker='o', alpha=0.5)
        ax_diff.set_xticks([0, 1])
        ax_diff.set_xticklabels(['Known', 'Unknown'], fontsize=12)
        ax_diff.set_ylabel('Rating (1-7)', fontsize=14)
        ax_diff.set_title(f'Participant Ratings in {construct.replace("_", " ")}', fontsize=16)
        ax_diff.tick_params(axis='x', labelsize=12)
        max_rating = max(df_constructs[f'{construct}_Known'].max(), df_constructs[f'{construct}_Unknown'].max())
        annotation_y = max_rating + 0.5
        ax_diff.text(0.5, annotation_y, significance_text, ha='center', va='bottom', fontsize=12, fontweight='bold')
        ax_diff.set_ylim(0, annotation_y + 1)
        plot_filename = os.path.join(output_dir, f"{construct}_Boxplot_and_Paired_Differences.png")
        plt.savefig(plot_filename, bbox_inches='tight', dpi=300)
        print(f"{construct} boxplot saved to {plot_filename}")
        plt.tight_layout()
        plt.show()

    # Combined distributions for Awareness, Fatigue, and Adjust Grasp
    fig, axes = plt.subplots(1, 3, figsize=(18, 6))

    sns.boxplot(y=df_constructs['Awareness'], ax=axes[0], color="skyblue")
    sns.swarmplot(y=df_constructs['Awareness'], color=".25", size=7, ax=axes[0])
    axes[0].set_title('Awareness Distribution', fontsize=20)  # Increased font size
    axes[0].set_ylabel('Awareness Rating (1-7)', fontsize=16)
    axes[0].tick_params(axis='y', labelsize=14)

    sns.boxplot(y=df_constructs['Fatigue'], ax=axes[1], color="skyblue")
    sns.swarmplot(y=df_constructs['Fatigue'], color=".25", size=7, ax=axes[1])
    axes[1].set_title('Fatigue Distribution', fontsize=20)  # Increased font size
    axes[1].set_ylabel('Fatigue Rating (1-7)', fontsize=16)
    axes[1].tick_params(axis='y', labelsize=14)

    sns.boxplot(y=df_constructs['Adjust Grasp'], ax=axes[2], color="skyblue")
    sns.swarmplot(y=df_constructs['Adjust Grasp'], color=".25", size=7, ax=axes[2])
    axes[2].set_title('Adjust Grasp Distribution', fontsize=20)  # Increased font size
    axes[2].set_ylabel('Adjust Grasp Rating (1-7)', fontsize=16)
    axes[2].tick_params(axis='y', labelsize=14)

    plt.suptitle('Distributions of Awareness, Fatigue, and Adjust Grasp', fontsize=24)
    plt.tight_layout(rect=[0, 0.03, 1, 0.95])

    # Save the combined distributions plot
    combined_plot_path = os.path.join(output_dir, "Combined_Awareness_Fatigue_AdjustGrasp.png")
    plt.savefig(combined_plot_path, bbox_inches='tight', dpi=300)
    print(f"Combined distributions plot saved to {combined_plot_path}")

    plt.show()

    # One-sample Wilcoxon tests for Awareness, Fatigue, Adjust Grasp
    neutral = 4  # neutral midpoint on Likert scale
    test_results = {}
    for construct in ['Awareness', 'Fatigue', 'Adjust Grasp']:
        stat, p_value = wilcoxon(df_constructs[construct] - neutral)
        test_results[construct] = (stat, p_value)
        print(f"One-sample Wilcoxon test for {construct}: Statistic={stat:.4f}, p-value={p_value:.4f}")

    for construct, (stat, p_value) in test_results.items():
        if p_value < 0.05:
            print(f"{construct}: Median significantly differs from neutral (p={p_value:.4f}).")
        else:
            print(f"{construct}: No significant difference from neutral median (p={p_value:.4f}).")

def correlation_analysis(df_constructs, constructs, output_dir):
    construct_cols = list(constructs.keys())
    corr_matrix = df_constructs[construct_cols].corr(method='spearman')
    plt.figure(figsize=(10, 5))
    sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f', square=False,
                linewidths=0.5, cbar_kws={"shrink": 0.5}, vmin=-1, vmax=1)
    plt.title('Spearman Correlation Matrix of Constructs', fontsize=16)
    plt.xticks(rotation=30, ha='right', fontsize=12)
    plt.yticks(fontsize=12)
    plt.tight_layout()
    corr_heatmap_path = os.path.join(output_dir, "Constructs_Correlation_Heatmap.png")
    plt.savefig(corr_heatmap_path, bbox_inches='tight')
    print(f"Correlation heatmap saved to {corr_heatmap_path}")
    plt.show()

def visualize_responses(df, df_Statements, reversed, output_dir):
    df_long = df.melt(id_vars='participant_id', 
                      value_vars=[f'Q{i}' for i in range(1, 18)],
                      var_name='Statement', 
                      value_name='Response')
    df_long = df_long.merge(df_Statements, left_on='Statement', right_index=True)
    df_pivot = df_long.pivot(index='participant_id', columns='Statement', values='Response')
    df_pivot = df_pivot[sorted(df_pivot.columns, key=lambda x: int(x[1:]))]
    fig = plt.figure(figsize=(15.7, 9))
    ax = sns.heatmap(
        df_pivot,
        annot=True,
        fmt='.0f',
        annot_kws={'size': 14},
        cmap='coolwarm',
        cbar_kws={'label': 'Response'},
        linewidths=0.5,
        linecolor='gray'
    )
    colorbar = ax.collections[0].colorbar
    colorbar.ax.set_ylabel(colorbar.ax.get_ylabel(), fontsize=18)
    colorbar.ax.tick_params(labelsize=14)
    pos = ax.get_position()
    ax.set_position([pos.x0, pos.y0 + 0.2, pos.width, pos.height])
    if reversed:
        plt.title("Participants' Reversed Responses to Each Statement", fontsize=21)
        filename_suffix = "Reversed_Responses"
    else:
        plt.title("Participants' Responses to Each Statement", fontsize=21)
        filename_suffix = "Original_Responses"
    plt.xlabel('Statements', fontsize=17)
    plt.ylabel('Participants', fontsize=17)
    plt.yticks(rotation=0, fontsize=14)
    plt.xticks(rotation=0, fontsize=14)
    statement_texts = [f"{index}: {row['Statement_Text']}" for index, row in df_Statements.iterrows()]
    all_text = "\n".join(statement_texts)
    plt.gcf().text(0.00, -0.2, all_text, fontsize=18, ha='left', va='top', 
                   transform=plt.gca().transAxes)
    plt.tight_layout(rect=[0, 0.3, 1, 1])
    responses_heatmap_path = os.path.join(output_dir, f"Survey_Responses_{filename_suffix}.png")
    plt.savefig(responses_heatmap_path, bbox_inches='tight', dpi=300)
    print(f"Survey responses heatmap ({filename_suffix}) saved to {responses_heatmap_path}")
    plt.show()

def main():
    output_dir = r"C:\Users\schmi\Documents\Studium\TUM\5. Semester\Masterthesis\Results Statistical Analysis\Q4 Survey Analysis"
    os.makedirs(output_dir, exist_ok=True)
    print(f"Plots will be saved to: {output_dir}")
    polycor = importr('polycor')
    pandas2ri.activate()
    reverse_coded_pairs = [
        ("Q1", "Q9"),
        ("Q2", "Q6"),
        ("Q10", "Q14"),
        ("Q3", "Q8"),
        ("Q16", "Q11"),
        ("Q4", "Q13"),
        ("Q5", "Q15"),
        ("Q7", "Q12"),
    ]
    exclude_pairs = [("Q7", "Q12")]
    try:
        data = load_participant_data(base_path)
        if not data:
            raise ValueError("No participant data found.")
        df_constructs, df_consistency, df_original, df_Statements = process_data(data)
        df_constructs_mean, constructs = compute_constructs(df_constructs)
        print("\nReliability Measures for Each Construct:")
        for construct, Statements in constructs.items():
            if len(Statements) > 1:
                itemscores = df_constructs[[f'Q{i}' for i in Statements]]
                polychoric_mat = compute_polychoric_matrix(itemscores)
                ord_alpha = ordinal_alpha(polychoric_mat, itemscores)
                omega = mcdonald_omega(itemscores)
                print(f"{construct}: Ordinal Alpha = {ord_alpha:.2f}, Omega = {omega:.2f}")
            else:
                print(f"{construct}: Only one item, reliability metrics not applicable.")
        analyze_data(df_constructs_mean, constructs, output_dir)
        visualize_responses(df_original, df_Statements, reversed=False, output_dir=output_dir)
        visualize_responses(df_constructs, df_Statements, reversed=True, output_dir=output_dir)
        df_consistency_scores, avg_consistency_scores, var_consistency_scores = compute_consistency_score(
            df_consistency, reverse_coded_pairs, exclude_pairs)
        plot_consistency_table(df_consistency_scores, avg_consistency_scores, var_consistency_scores, output_dir)
        csv_path = os.path.join(output_dir, "participant_consistency_scores.csv")
        df_consistency_scores.to_csv(csv_path, index=False)
        print(f"Participant consistency scores saved to {csv_path}")
    except Exception as e:
        print("An error occurred:", e)
        import traceback
        traceback.print_exc()

if __name__ == "__main__":
    main()
```
